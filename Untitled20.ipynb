```python
# --- Common Setup for All Tasks (Fully Fixed for Latest Qiskit) ---
!pip install qiskit qiskit-aer --quiet

from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from qiskit_aer.noise import NoiseModel, depolarizing_error  # âœ… Fixed import

# Define Grover oracle
def grover_oracle(marked_state):
    """Oracle that flips the phase of the marked state."""
    n = len(marked_state)
    qc = QuantumCircuit(n)
    for i, bit in enumerate(marked_state):
        if bit == "0":
            qc.x(i)
    qc.h(n-1)
    qc.mcx(list(range(n-1)), n-1)
    qc.h(n-1)
    for i, bit in enumerate(marked_state):
        if bit == "0":
            qc.x(i)
    return qc

# Define diffuser
def diffuser(n):
    qc = QuantumCircuit(n)
    qc.h(range(n))
    qc.x(range(n))
    qc.h(n-1)
    qc.mcx(list(range(n-1)), n-1)
    qc.h(n-1)
    qc.x(range(n))
    qc.h(range(n))
    return qc

print("âœ… Setup complete: Oracle and Diffuser ready for all Grover tasks.")
```

```python
# --- Task 1: Change the Marked State ---
marked_states = ["001", "010", "111"]

for marked_state in marked_states:
    n = len(marked_state)
    qc = QuantumCircuit(n, n)
    qc.h(range(n))
    qc.append(grover_oracle(marked_state), range(n))
    qc.append(diffuser(n), range(n))
    qc.measure(range(n), range(n))

    sim = Aer.get_backend('aer_simulator')
    compiled = transpile(qc, sim)
    result = sim.run(compiled, shots=1024).result()
    counts = result.get_counts()

    print(f"\nğŸ”¹ Marked state: {marked_state}")
    print(qc.draw('text'))
    plot_histogram(counts)
    plt.show()
```

```python
# --- Task 2: 4-Qubit Grover Search ---
n = 4
marked_state = "1010"

qc = QuantumCircuit(n, n)
qc.h(range(n))
qc.append(grover_oracle(marked_state), range(n))
qc.append(diffuser(n), range(n))
qc.measure(range(n), range(n))

sim = Aer.get_backend('aer_simulator')
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()
counts = result.get_counts()

print(qc.draw('text'))
plot_histogram(counts)
plt.show()
```

```python
# --- Task 3: Multiple Iterations ---
n = 3
marked_state = "011"

qc = QuantumCircuit(n, n)
qc.h(range(n))
for _ in range(2):  # run oracle + diffuser twice
    qc.append(grover_oracle(marked_state), range(n))
    qc.append(diffuser(n), range(n))
qc.measure(range(n), range(n))

sim = Aer.get_backend('aer_simulator')
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()
counts = result.get_counts()

print(qc.draw('text'))
plot_histogram(counts)
plt.show()
```

```python
# --- Task 5: Experiment with Noise ---
from qiskit_aer.noise import NoiseModel, depolarizing_error

n = 3
marked_state = "101"

# Create a simple depolarizing noise model (1% error on single-qubit gates)
noise_model = NoiseModel()
error_1q = depolarizing_error(0.01, 1)
error_2q = depolarizing_error(0.02, 2)
noise_model.add_all_qubit_quantum_error(error_1q, ['u1', 'u2', 'u3'])
noise_model.add_all_qubit_quantum_error(error_2q, ['cx'])

# Create Grover circuit
qc = QuantumCircuit(n, n)
qc.h(range(n))
qc.append(grover_oracle(marked_state), range(n))
qc.append(diffuser(n), range(n))
qc.measure(range(n), range(n))

# Simulators
sim_ideal = Aer.get_backend('aer_simulator')
sim_noisy = Aer.get_backend('aer_simulator')

# Run ideal
compiled_ideal = transpile(qc, sim_ideal)
ideal_result = sim_ideal.run(compiled_ideal, shots=1024).result()
counts_ideal = ideal_result.get_counts()

# Run noisy
compiled_noisy = transpile(qc, sim_noisy)
noisy_result = sim_noisy.run(compiled_noisy, shots=1024, noise_model=noise_model).result()
counts_noisy = noisy_result.get_counts()

# Compare
print("âœ… Ideal vs Noisy Simulation Results")
plot_histogram([counts_ideal, counts_noisy], legend=["Ideal", "Noisy"])
plt.show()
```
